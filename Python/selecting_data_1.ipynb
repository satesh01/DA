# Selecting Data

```python
import pandas as pd
```

```python
countries_data = pd.read_csv('top_10_countries.csv')
```

### Selecing a single column

```python
countries_data
```

```python
# Selecing the 'Region' column
countries_data['Region']
```

```python
# A single column is a Series object - a DataFrame is a collection of Series objects
type(countries_data['Region'])
```

### Selecting multiple columns

```python
countries_data
```

```python
# multiple columns must be passed into a list
countries_data[['Region','Country / Dependency', 'Rank']]
```

### Selecting rows and columns

```python
countries_data
```

```python
# iloc requires index positions
countries_data.iloc[3]
```

```python
# Slicing the rows
countries_data.iloc[3:]
```

```python
# Reversing the rows
countries_data.iloc[::-1]
```

```python
countries_data
```

```python
# indexing rows and columns
countries_data.iloc[2,3]
```

```python
# Slicing rows and columns
countries_data.iloc[2:,1:4]
```

```python
# loc requires label names, it does not accept index positions
countries_data.loc[2,1]
```

```python
countries_data
```

```python
# slicing rows and columns with loc requires label names
countries_data.loc[2:,['Country / Dependency','Population']]
```

### Conditional Selection

```python
countries_data
```

```python
# Returns True if any value equals 'Asia'
countries_data=='Asia'
```

```python
# Only populates a value in the array if it is 'Asia'
countries_data[countries_data=='Asia']
```

```python
# Returns True if the 'Region' column contains the value 'Asia'
countries_data['Region']=='Asia'
```

```python
# Returns the rows only if the 'Region' column contains 'Asia'
countries_data[countries_data['Region']=='Asia']
```

```python
# You can combine multiple conditions by enclosing them in parenthesis
# for logical operations where all conditions must be True use & - this is the AND logical operator sign
countries_data[(countries_data['Region']=='Asia') & (countries_data['Population']>300000000)]
```

```python
# You can combine multiple conditions by enclosing them in parenthesis
# for logical operations where all conditions must be True use | - this is the OR logical operator sign
countries_data[(countries_data['Region']=='Asia') | (countries_data['Population']>300000000)]
```

```python
# You can stack on additional filters
countries_data[(countries_data['Region']=='Asia') | (countries_data['Population']>300000000)][['Rank','Country / Dependency']]
```

### head() and tail()

```python
# head returns the first 5 rows of a DataFrame
countries_data.head()
```

```python
# tail returns the last 5 rows of a DataFrame
countries_data.tail()
```
